```python
import glob
import pandas as pd
import geopandas as gpd
import fnmatch
import os
import matplotlib.pyplot as plt
from shapely.geometry import Point
from geopandas import GeoDataFrame
import numpy as np
```

```python
airports = pd.read_csv('airportstable2.csv')
```

```python
airports = airports.replace(r'^\s*$', np.NaN, regex=True)
```

```python
internationalairports = airports[airports['AIRPORT_AÉROPORT'].str.contains('International')]
```

```python
internationalairports
```

```python
from geopandas import GeoDataFrame
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(internationalairports.LONGITUDE, internationalairports.LATTITUDE)]
df = internationalairports.drop(['LONGITUDE', 'LATTITUDE'], axis=1)
airportgdf = GeoDataFrame(df, crs="EPSG:3347", geometry=geometry)
```

```python
airportgdf.to_file('internationalairports.geojson', driver='GeoJSON')  
```

```python
healthfacility = pd.read_csv('healthfacilitydata.csv')
```

```python
healthfacility = healthfacility.replace(r'^\s*$', np.NaN, regex=True)
```

```python
hospitals = healthfacility[healthfacility['odhf_facility_type'].str.contains('Hospitals')]
```

```python
hospitals['source_facility_type'] = hospitals['source_facility_type'].astype(str)
```

```python
genhospitals = hospitals[hospitals['source_facility_type'].str.contains('general')]
genhospitals
```

```python
from geopandas import GeoDataFrame
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(genhospitals.longitude, genhospitals.latitude)]
df = genhospitals.drop(['longitude', 'latitude'], axis=1)
gdf = GeoDataFrame(df, crs="EPSG:3347", geometry=geometry)
```

```python
gdf.to_file('generalhospitals.geojson', driver='GeoJSON')  
```
